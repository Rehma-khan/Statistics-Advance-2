Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

Ans. Type I and Type II Errors in Hypothesis Testing

When we do hypothesis testing, we start with a Null Hypothesis (H‚ÇÄ) that we assume is true, and then we check if there is enough evidence to reject it.
While doing this, there are chances of making mistakes. These mistakes are called Type I and Type II errors.

**1. Type I Error (False Positive)**

* This error occurs when we reject the null hypothesis (H‚ÇÄ) even though it is actually true.

* In simple words: We think something is happening, but in reality, nothing is happening.

**2. Type II Error (False Negative)**

* This error happens when we fail to reject the null hypothesis even though it is false.

* In simple words: Something is happening, but we miss it.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

Ans. **P-value in Hypothesis Testing**

The P-value is the probability that the results we observed (or something more extreme) could happen just by chance, assuming the null hypothesis (H‚ÇÄ) is true.

In simple words:
P-value tells us how strong the evidence is against the null hypothesis.


**How to Interpret the P-value**

* If P-value is small (usually ‚â§ 0.05):

It means the result is not likely due to chance.

So, we reject the null hypothesis (H‚ÇÄ).

There is strong evidence against H‚ÇÄ.

* If P-value is large (greater than 0.05):

It means the result could be due to chance.

So, we do not reject the null hypothesis (H‚ÇÄ).

There is weak evidence against H‚ÇÄ.

Question 3:Explain the difference between a Z-test and a T-test, including when to use
each.

Ans. **Difference Between Z-test and T-test**

Both Z-test and T-test are used in hypothesis testing to compare sample data with population data, but they are used in different conditions.

**1. Z-test**

* Used when sample size is large (usually n > 30).

* Used when population standard deviation (œÉ) is known.

* Test is based on the Z-distribution (normal distribution).

**2. T-test**

* Used when sample size is small (usually n ‚â§ 30).

* Used when population standard deviation (œÉ) is unknown.

* Test is based on the t-distribution, which has thicker tails than the normal distribution.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

Ans. **Purpose of ANOVA**

ANOVA (Analysis of Variance) is used to compare the means of three or more groups to check whether they are significantly different from each other.

It helps answer the question:
‚ÄúAre these groups different, or are the differences just due to random chance?‚Äù

**Assumptions of ANOVA**

To use ANOVA correctly, these conditions must be satisfied:

1. Independence:
Observations in each group must be independent of each other.

2. Normality:
Data in each group should be approximately normally distributed.

3. Equal Variances (Homogeneity of Variance):
All groups should have similar or equal variances.

**How ANOVA Extends Hypothesis Testing**

* When there are only two groups, we can use a t-test.

* But when there are three or more groups, doing multiple t-tests increases the chance of errors.

* ANOVA solves this by testing all groups at once using one test.

Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Sample data (example dataset)
data = [52, 55, 49, 50, 48, 53, 51, 54, 50, 49]

# Given information
population_mean = 50       # Hypothesized mean (H0)
population_std = 3         # Known population standard deviation
alpha = 0.05               # Significance level

# Step 1: Calculate sample mean
sample_mean = np.mean(data)

# Step 2: Calculate Z-statistic
n = len(data)
z_value = (sample_mean - population_mean) / (population_std / sqrt(n))

# Step 3: Calculate P-value
p_value = 1 - norm.cdf(abs(z_value))  # two-tailed test
p_value = p_value * 2

# Print results
print("Sample Mean =", sample_mean)
print("Z-value =", z_value)
print("P-value =", p_value)

# Step 4: Decision
if p_value < alpha:
    print("Reject the Null Hypothesis (H0). The sample mean is significantly different.")
else:
    print("Fail to Reject the Null Hypothesis (H0). The sample mean is not significantly different.")
```

Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate multiple samples from a non-normal distribution (Exponential)
np.random.seed(42)
sample_size = 1000
n_samples = 1000
data = np.random.exponential(scale=1.0, size=(n_samples, sample_size))

# Compute sample means to demonstrate the Central Limit Theorem
sample_means = np.mean(data, axis=1)

# Plot histogram of sample means
plt.figure()
plt.hist(sample_means, bins=30)
plt.xlabel('Sample Means')
plt.ylabel('Frequency')
plt.title('Central Limit Theorem Demonstration')
plt.show()
```

Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    # Calculate sample mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    standard_error = sample_std / np.sqrt(n)

    # Calculate Z critical value
    z_value = norm.ppf((1 + confidence) / 2)

    # Confidence interval
    margin_of_error = z_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print results
    print("Sample Mean:", sample_mean)
    print(f"{int(confidence*100)}% Confidence Interval:", (lower_bound, upper_bound))

    # Plot data and confidence interval
    plt.figure()
    plt.hist(data, bins=20)
    plt.axvline(sample_mean)
    plt.axvline(lower_bound)
    plt.axvline(upper_bound)
    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.title('Confidence Interval Visualization')
    plt.show()

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=5, size=100)
plot_confidence_interval(sample_data)
```
